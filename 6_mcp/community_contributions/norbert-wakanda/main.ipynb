```python
from mcp_agent import search_known_issues, log_terminal_activity
from IPython.display import Markdown, display

import asyncio
import nest_asyncio
```

```python
import gradio as gr

nest_asyncio.apply()
```

```python
import time

async def check_and_log_terminal_issues():
    """
    Check the log file for any issues and log them to the database,
    then display the result as Markdown output.
    """
    request = "Check the log file for any issues and log them to the database"
    result = await log_terminal_activity(request)
    display(Markdown(result))

# The following periodic log checker runs every 1 minute (60 seconds).
# You can adjust the interval by changing the argument to periodic_log_check.

async def periodic_log_check(interval_minutes=10000):
    """
    Run check_and_log_terminal_issues() every `interval_minutes`.
    By default, this runs every 1 minute.
    """
    while True:
        await check_and_log_terminal_issues()
        await asyncio.sleep(interval_minutes * 60)

# Start the periodic task in the event loop
# This will trigger the check every 1 minute.
loop = asyncio.get_event_loop()
task = loop.create_task(periodic_log_check(100000))
```

```python


async def query_known_issues(request):
    """"function to search for known issues"""
    result = await search_known_issues(request)
    return result

iface = gr.Interface(
    fn=query_known_issues,
    inputs=gr.Textbox(lines=2, placeholder="Describe the issue..."),
    outputs=gr.Markdown(),
    title="Search Known Issues",
    description="Enter an issue description to search known/previously solved issues."
)

iface.launch()
```
