```python
from __future__ import annotations

import sys
from pathlib import Path

from dotenv import load_dotenv


def find_bernard_mcp_root() -> Path:
    p = Path.cwd().resolve()
    for _ in range(24):
        candidate = p / "6_mcp" / "community_contributions" / "BernardUdo"
        if (candidate / "server.py").is_file():
            return candidate
        if p.name == "BernardUdo" and (p / "server.py").is_file():
            return p
        if p == p.parent:
            break
        p = p.parent
    raise RuntimeError(
        "Could not find 6_mcp/community_contributions/BernardUdo/server.py. "
        "cd to repo root or the BernardUdo folder."
    )


MCP_ROOT = find_bernard_mcp_root()
if str(MCP_ROOT) not in sys.path:
    sys.path.insert(0, str(MCP_ROOT))

from winpatch import winpatch_mcpserver_stdio

winpatch_mcpserver_stdio()
load_dotenv(override=True)
pass
```

## List tools (stdio MCP)

If `list_tools` fails, use `await server.session.list_tools()` (Week 6 lab note).


```python
from agents.mcp import MCPServerStdio

mcp_params = {
    "command": "uv",
    "args": ["run", "server.py"],
    "cwd": str(MCP_ROOT),
}


async def list_mcp_tools():
    async with MCPServerStdio(
        params=mcp_params, client_session_timeout_seconds=60
    ) as server:
        try:
            out = await server.list_tools()
        except TypeError:
            out = await server.session.list_tools()
        if hasattr(out, "tools"):
            out = out.tools
        return out


tool_list = await list_mcp_tools()
print(len(tool_list), [t.name for t in tool_list])
```
