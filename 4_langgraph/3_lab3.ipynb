### Asynchronous LangGraph

To run a tool:  
Sync: `tool.run(inputs)`  
Async: `await tool.arun(inputs)`

To invoke the graph:  
Sync: `graph.invoke(state)`  
Async: `await graph.ainvoke(state)`

````python
class State(TypedDict):
    
    messages: Annotated[list, add_messages]


graph_builder = StateGraph(State)
````

````python
pushover_token = os.getenv("PUSHOVER_TOKEN")
pushover_user = os.getenv("PUSHOVER_USER")
pushover_url = "https://api.pushover.net/1/messages.json"

def push(text: str):
    """Send a push notification to the user"""
    requests.post(pushover_url, data = {"token": pushover_token, "user": pushover_user, "message": text})

tool_push = Tool(
        name="send_push_notification",
        func=push,
        description="useful for when you want to send a push notification"
    )
````

````python
all_tools = tools + [tool_push]
````

````python

llm = ChatOpenAI(model="gpt-4o-mini")
llm_with_tools = llm.bind_tools(all_tools)


def chatbot(state: State):
    return {"messages": [llm_with_tools.invoke(state["messages"])]}
````

````python
# Ahora podemos crear el gráfico, que es una representación del flujo de nuestra aplicación. 
# Añadimos nodos, que son las funciones que queremos ejecutar, y aristas, que son las conexiones entre los nodos. 
# También podemos añadir aristas condicionales, que solo se ejecutan si se cumple una determinada condición. 
# En este caso, queremos ejecutar el nodo «tools» si se cumple la condición «tools_condition», 
# que comprueba si el último mensaje del chatbot contiene una llamada a una herramienta. 
# Por último, compilamos el gráfico y lo mostramos.


graph_builder = StateGraph(State) # crear un generador de gráficos con el tipo de estado que hemos definido
graph_builder.add_node("chatbot", chatbot) # agregar un nodo al gráfico, con el nombre "chatbot" y la función chatbot que definimos antes
graph_builder.add_node("tools", ToolNode(tools=all_tools)) #
graph_builder.add_conditional_edges( "chatbot", tools_condition, "tools")
graph_builder.add_edge("tools", "chatbot")
graph_builder.add_edge(START, "chatbot")

memory = MemorySaver()
graph = graph_builder.compile(checkpointer=memory)
display(Image(graph.get_graph().draw_mermaid_png()))
````

````python
config = {"configurable": {"thread_id": "10"}}

async def chat(user_input: str, history):
    result = await graph.ainvoke({"messages": [{"role": "user", "content": user_input}]}, config=config)
    return result["messages"][-1].content


gr.ChatInterface(chat, type="messages").launch()
````
