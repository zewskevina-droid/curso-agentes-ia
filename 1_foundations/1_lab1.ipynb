```python
# Verificar si la clave API de OpenAI está configurada
import os
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
# Importar OpenAI y crear cliente
import openai
openai = OpenAI()
```

```python
# Crear una lista de mensajes en el formato familiar de OpenAI
messages = [{"role": "user", "content": "What is 2+2?"}]
```

```python
# Y ahora llámalo! Cualquier problema, ve a la guía de resolución de problemas
# Esto usa GPT 4.1 nano, el modelo increíblemente barato
# La guía de APIs (guía 9) tiene instrucciones exactas para usar alternativas aún más baratas o gratuitas a OpenAI
# Si obtienes un NameError, ve a la carpeta de guías (guía 6) para aprender sobre NameErrors - siempre solucionable al instante

response = openai.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages
)

print(response.choices[0].message.content)
```

```python
# Y ahora - preguntemos por una pregunta:

question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
# Preguntarlo - esto usa GPT 4.1 mini, aún barato pero más poderoso que nano

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
# ask it - this uses GPT 4.1 mini, still cheap but more powerful than nano

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

question = response.choices[0].message.content

print(question)
```
